# Planets DataSet
Seaborn’un planets veri seti, ötegezegen (güneş sistemi dışındaki gezegenler) keşifleri hakkında bilgiler içeren, gökbilimle ilgili bir örnek veri setidir. Veri seti hem istatistiksel analiz hem de görselleştirme pratikleri için sıkça kullanılır.

***Genel Bilgi***

***Toplam satır sayısı:*** 1035

***Toplam sütun sayısı:*** 6

***Sütunlar ve Açıklamaları***

***method*** → Ötegezegenin keşfedilme yöntemi

Örnek: Radial Velocity, Transit, Imaging, Eclipse Timing Variations

Yani gezegenin varlığını tespit etmek için kullanılan astronomik teknik.

***number*** → Sistem içindeki gezegen sayısı

Bir yıldız etrafında kaç gezegen keşfedildiğini gösterir.

Genelde 1, ama daha fazla olanlar da var.

***orbital_period*** → Gezegenin yörünge periyodu (gün cinsinden)

Gezegenin yıldızı etrafında tam tur atma süresi.

Bazıları birkaç gün, bazıları yüzlerce gün sürebilir.

***mass*** → Gezegenin kütlesi (Jüpiter kütlesi birimiyle)

Bazı değerler eksik (NaN).

Büyük gezegenler genelde Jüpiter büyüklüğünde veya daha ağırdır.

***distance*** → Dünya’dan uzaklığı (parsec cinsinden)

1 parsec ≈ 3.26 ışık yılı.

Gezegenlerin bize ne kadar uzak olduğunu gösterir.

***year*** → Keşif yılı

1989’dan 2014’e kadar uzanır.

Zaman içinde keşiflerdeki artışı görselleştirmek için kullanılır.

***Veri Setinin Kullanım Amaçları***

Keşif yöntemlerinin yıllara göre dağılımını incelemek.

Kütle, yörünge periyodu ve uzaklık gibi özelliklerin istatistiksel dağılımlarını görmek.

Hangi yöntemlerin hangi tip gezegenleri bulmada daha etkili olduğunu analiz etmek.

Zaman serisi analizleri (ör. yıllara göre keşif sayısı).
