```python
plt.figure(figsize = (8,6))

plt.scatter(tips['total_bill'],tips['tip'])

plt.xlabel('total_bill',fontsize = 14)
plt.ylabel('tips',fontsize = 14)
plt.title('Toplam Tutar ve Bahşiş Durumu', fontsize = 18)

plt.show()
```
